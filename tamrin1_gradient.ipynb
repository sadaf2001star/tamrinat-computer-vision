```python
import numpy as np
import matplotlib.pyplot as plt
width, height = 256, 256
# نقاط شروع و پایان
point1 = (220, 150) # نقطه اول (x1, y1)
point2 = (90, 250) # نقطه دوم (x2, y2)
#یک تصویر سیاه
image = np.zeros((height, width, 3), dtype=np.uint8)
color1 = np.array([255,0, 0])
color2 = np.array([0,0, 255])
#  گرادیانت
for x in range(width):
 for y in range(height):
  gd = np.linalg.norm(np.array([x, y]) - np.array(point1))
  hh = np.linalg.norm(np.array([x, y]) - np.array(point2))
  total_distance = gd + hh
  if total_distance > 0:
   ratio = gd/ total_distance
   color = (1 - ratio) * color1 + ratio * color2
   image[y, x] = color.astype(np.uint8)
# نمایش
plt.imshow(image)
plt.axis('off')
plt.show()
```
